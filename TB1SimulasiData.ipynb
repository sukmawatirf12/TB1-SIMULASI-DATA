```python
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_csv('employee.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df_copy = df[:]
df_copy['Attrition'] = le.fit_transform(df_copy['Attrition'])
corr = df_copy.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
cols_to_drop = ['Unnamed: 0', 'EmployeeNumber', 'DailyRate', 'Education', 'EnvironmentSatisfaction', 
                'HourlyRate', 'JobInvolvement', 'JobSatisfaction', 'MonthlyRate', 'NumCompaniesWorked', 'PercentSalaryHike',
                'PerformanceRating', 'RelationshipSatisfaction', 'StandardHours', 'TrainingTimesLastYear', 'WorkLifeBalance', 
                'YearsSinceLastPromotion', 'Over18']
df = df.drop(columns=cols_to_drop)
```

```python
df
```

```python
y = df['Attrition']
X = df.drop(columns=['Attrition'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
```

```python
X_train_encoded = pd.get_dummies(X_train, columns=['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'OverTime'])
X_test_encoded = pd.get_dummies(X_test, columns=['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'OverTime'])

print(X_train_encoded.head())
print(X_test_encoded.head())
```

```python
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_final = scaler.fit_transform(X_train_encoded)

X_test_final = scaler.fit_transform(X_test_encoded)



print(X_train_final.shape)
print(X_test_final.shape)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)

scores = cross_val_score(model, X_train_final, y_train, cv=5)

print('Cross-validation scores:', scores)
print('Mean cross-validation score:', scores.mean())
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report
estimator = SVC()
estimator.fit(X_train_final, y_train)

# Evaluate estimator on trainset
train_preds = estimator.predict(X_train_final)
train_report = classification_report(y_train, train_preds)
print("Classification report for trainset:\n", train_report)

# Evaluate estimator on testset
test_preds = estimator.predict(X_test_final)
test_report = classification_report(y_test, test_preds)
print("Classification report for testset:\n", test_report)
```
